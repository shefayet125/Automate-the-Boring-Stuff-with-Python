# <b><b> Useful String Methods

## capitalize()

```python
#The capitalize() method returns a string where the first character is upper case, and the rest is lower case.
s= 'This Is a Beautiful Day'
s.capitalize()
```

## casefold() & lower() & upper()

The `casefold()` method returns a string where all the characters are `lower` case.

This method is similar to the `lower()` method, but the casefold() method is stronger, more aggressive, casefold() is more aggressive than lower(). While lower() only converts ASCII characters to lowercase, casefold() converts both ASCII and non-ASCII characters to lowercase.

```python
ss = 'Shefayet'
print(ss.casefold())
print(ss.lower())
print(ss.upper())
```

```python
string = "ß"
print(f"Original string: {string}")
print(f"Converted using casefold(): {string.casefold()}")
print(f"Converted using lower(): {string.lower()}")
```

## center()

```python
'''The center() method will center align the string,
using a specified character (space is default) as the fill character.'''

sh = 'Shefayet'
sh.center(20,'*')
```

```python
#The count() method returns the number of times a specified value appears in the string.
c ='You are a good boy,You made good impressions.'
c.count('good')
```

```python
c.count('o',10,20)
```

```python
c.count('a',0,22)
```

## encode()

```python
#The encode() method encodes the string, using the specified encoding.Default is UTF-8 will be used.
txt = "My name is Ståle"

print(txt.encode(encoding="ascii",errors="backslashreplace"))
print(txt.encode(encoding="ascii",errors="ignore"))
print(txt.encode(encoding="ascii",errors="namereplace"))
print(txt.encode(encoding="ascii",errors="replace"))
print(txt.encode(encoding="ascii",errors="xmlcharrefreplace"))
```

## endswith() & startswith()

```python
#The endswith() method returns True if the string ends with the specified value, otherwise False.
txt = "Hello, welcome to my world."
x = txt.endswith("my world.")
print(x)
```

```python
txt = "Hello, welcome to my world."
x = txt.endswith("my world.",5,11)
print(x)
```

```python
txt = "Hello, welcome to my world."
x = txt.startswith("Hello")
print(x)
```

## format()

```python
txt = "For only {price:.2f} dollars!"
print(txt.format(price = 49))
```

```python
#The placeholders can be identified using named indexes {price}, numbered indexes {0}, or even empty placeholders {}.

txt1 = "My name is {fname}, I'm {age}".format(fname = "John", age = 36)
txt2 = "My name is {0}, I'm {1}".format("John",36)
txt3 = "My name is {}, I'm {}".format("John",36)
print(txt1)
print(txt2)
print(txt3)
```
