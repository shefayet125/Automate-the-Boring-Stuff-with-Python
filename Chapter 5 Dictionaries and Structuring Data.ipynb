# Dictionaries vs. Lists

Unlike lists, items in dictionaries are unordered. The first item in a list 
named spam would be bird[0]. But there is no “first” item in a dictionary. 
While the order of items matters for determining whether two lists are the 
same, it does not matter in what order the key-value pairs are typed in a dictionary.

```python
bird=['parrot','sparrow','peacock','pigeon']
Bird=['pigeon','peacock','sparrow','parrot']
bird==Bird
```

```python
myFlower={'name':'rose','colour':'red','origin':'Australia','season':'summer','price':1000,'colour':'red'}
myflower==myFlower
```

# The keys(), values(), and items() Methods

There are three dictionary methods that will return list-like values of the 
dictionary’s keys, values, or both keys and values: keys(), values(), and items(). 
The values returned by these methods are not true lists: They cannot be 
modified and do not have an append() method. But these data types (dict_keys,dict_values, and dict_items, respectively) can be used in for loops.

```python
for s in myFlower.values():
    print(s)
```

```python
for s in myFlower.keys():
    print(s)
```

```python
myflower.items()
# item() is A view object that displays a list of a dictionary's key-value tuple pairs.
```

```python
# Iterate over the key-value pairs
for key, value in myFlower.items():
    print(key, value)
```

# The get() Method

```python
my_dict = {'name': 'John', 'age': 25, 'city': 'New York'}

name = my_dict.get('name')
print(name)  # Output: John

profession = my_dict.get('profession', 'Unknown')
print(profession)  # Output: Unknown (since 'profession' key doesn't exist in the dictionary).
```

# The setdefault() Method

The setdefault() method offers a way to do this in one line of code. The 
first argument passed to the method is the key to check for, and the second 
argument is the value to set at that key if the key does not exist. If the key 
does exist, the setdefault() method returns the key’s value

```python
blueeyes={'color':'blue','eyes':'blue'}
blueeyes.setdefault('hair','black')
```

```python
blueeyes.setdefault('skin','white')
```

```python
blueeyes.setdefault('eyes','grey')
```

```python
blueeyes
```

# The update() method

The update() method in Python dictionary is used to update the key-value pairs in a dictionary with the key-value pairs from another dictionary or an iterable of key-value pairs.

```python
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}

dict1.update(dict2)
print(dict1)
```
