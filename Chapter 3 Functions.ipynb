# <b><b> Return Values and return Statements

The return statement is used in programming languages to exit a function and return a value to the caller. When a return statement is executed, the function stops executing and control is returned to the code that called the function. The value specified in the return statement is the result of the function and can be used by the caller.

```python
def multiply(a,b):
    return a * b
```

```python
multiply(2,3)
```

```python
def subtraction(c,d):
    return c - d
```

```python
subtraction(22,10)
```

# <b><b>Local and Global Scope

## Local Variables Cannot Be Used in the Global Scope

```python
def age():
    year = 23
age()
```

```python
print(year)
```

<b><b>
* The error happens because the year variable exists only in the local scope created when age() is called.
* Once the program execution returns from age, that local scope is destroyed, and there is no longer a variable named year.
* So when your program tries to run print(year), Python gives you an error saying that year is not defined.
* This makes sense if you think about it; when the program execution is in the global scope, no local scopes exist, so there cant be any local variables. This is why only global variables can be used in the global scope.

## Local Scopes Cannot Use Variables in Other Local Scopes

```python
def spam():
    eggs = 99                #1
    bacon()                  #2
    print(eggs)              #3

def bacon():
    ham = 101
    eggs = 0                 #4

spam()                       #5
```

<b><b>
* When `spam()` calls the `bacon()` function, a new local scope is created for bacon(). Inside bacon(), the variables `ham` and `eggs` are local to bacon().
* Notice that `bacon()` has its own local eggs variable, which is completely separate from the eggs variable in spam(). Changing the value of eggs inside bacon() does `not affect` the eggs variable in spam().

## Local and Global Variables with the Same Name

```python
def spam():
 eggs = 'spam local' #1
 print(eggs) # prints 'spam local'

def bacon():
 eggs = 'bacon local' #2
 print(eggs) # prints 'bacon local'
 spam()
 print(eggs) # prints 'bacon local'

eggs = 'global'    #3
bacon()
print(eggs) # prints 'global'
```

# <b><b>Exception Handling

<b><b>
* Right now, getting an error, or exception, in your Python program means the entire program will crash. You dont want this to happen in real-world programs. Instead, you want the program to detect errors handle them, and then continue to run.

```python
def Division(Number):
    return 22/Number

print(Division(2))
print(Division(3))
print(Division(0))
print(Division(11))
```

```python
def Division(Number):
    try:
        return 22/Number
    except ZeroDivisionError:
        print('Error: Invalid argument.')

print(Division(2))
print(Division(3))
print(Division(0))
print(Division(11))
```

```python
def Division(Number):
    return 22/Number

try:
    print(Division(2))
    print(Division(3))
    print(Division(0))
    print(Division(11))
except ZeroDivisionError:
        print('Error: Invalid argument.')
```

<b><b>
* The reason print(Division(11)) is never executed is because once the execution jumps to the code in the except clause, it does not return to the try clause.
